```python
# Import the sys module for exit functionality
import sys 
```

```python
# Main Entry point of the program
def main():
    
    # Initialize the account balance
    balance=210000

    # Define the correct PIN as a string
    correct_pin=1234
    
    # Prompt the user to enter a 4-digit PIN
    pin=int(input('Please Enter your pin to continue '))
    
    # Check if the PIN entered is correct or not
    if pin == correct_pin:
        
        # If the PIN is correct, display the ATM menu
        print('''
            ****Welcome To Our ATM Project****
        
        Please Select Any One of the following
        1 Check Balance.
        2 Withdraw Amount.
        3 Deposit.
        4 Exit.
     ''')
    else:
        print("\nPin Entered is Wrong")
        
        # Exit the program if the PIN is invalid
        sys.exit("Wrong PIN entered. Exiting...")
    
    # Prompt the user to choose a transaction
    choose=int(input('Choose a transcation from above options '))
    
    if choose==1:
        print(f'\nAvailable account balance is {balance}')
    
    # Prompt the user to enter the amount for withdrawal
    elif choose==2:
        withdraw=int(input('\nEnter Amount for withdrawal '))
        
        # Deduct the withdrawal amount from the balance if sufficient funds are available
        if withdraw<=balance:
            balance-=withdraw
            print(f'''\n
             **** Withdrawal Successful ****
             
             Remaining Balance is {balance}
            
            ''')
        else:
                print('/n Insufficent Balance ')
  
    # Prompt the user to enter the amount for deposit     
    elif choose==3:
        Deposit=int(input('\nEnter Amount for Deposit '))
        
        # Add the deposit amount to the balance
        balance+=Deposit
        print(f'''\n
             **** Deposit Successful ****
             
             Available Balance is {balance}
            
            ''')
        
    # Exit the program if the user chooses to exit    
    elif choose == 4:
        sys.exit("Exiting...")

        
if __name__ == "__main__":
    main()
```
