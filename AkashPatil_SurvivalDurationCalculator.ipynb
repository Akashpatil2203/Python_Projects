```python
# Import the sys module for exit functionality
import sys
```

```python
# Main Entry point of the program
def main():
       
        print('''
            ****Welcome To Our Survial Duration Calculation Project****
     ''')
        
        # Prompt the user to enter their age
        Age=int(input('\nPlease Enter your Age to continue '))
        
        print('''\n
            ****Welcome To Our Survial Duration Calculation Project****

        Please Select Any One of the following
        1 Months.
        2 Weeks.
        3 Days.
        4 Hours.
        5 Minutes.
        6 Seconds.
        7 Exit.
        
     ''')
        
        # Calculate age in different units
        months = Age * 12
        weeks = Age * 52
        Days = Age * 365
        Hours = Days * 24
        Minutes = Hours * 60
        Seconds = Minutes * 60
   
        # Prompt the user to choose from options
        choose=int(input('Choose from above options '))
    
        # Calculate the user's age in the selected time unit
        if choose==1:
            months = Age * 12
           
            print(f'\nAge in Months is {months}')
            
        elif choose==2:
            weeks = Age * 52
            
            print(f'\nAge in weeks is {weeks}')
            
        elif choose==3:
            Days = Age * 365
            
            print(f'\nAge in Days is {Days}')
            
        elif choose==4:
            Hours = Days * 24
            
            print(f'\nAge in Hours is {Hours}')
            
        elif choose==5:
            Minutes = Hours * 60
            
            print(f'\nAge in Minutes is {Minutes}')
            
        elif choose==6:
            Seconds = Minutes * 60
            
            print(f'\nAge in Seconds is {Seconds}')
            
        elif choose==7:
            sys.exit("Exiting...")


        
if __name__ == "__main__":
    main()
```
